```python
from google_play_scraper import reviews, Sort
import pandas as pd
from datetime import datetime

apps = {
    "grab": "com.grabtaxi.passenger",  # Ganti dari "nujek"
    "maxim": "com.taxsee.taxsee",
    "gojek": "com.gojek.app"
}

# Batas tanggal: 1 April 2022
cutoff_date = datetime(2022, 4, 1)

def get_reviews(app_name, app_id, total=20000):
    all_reviews = []
    count = 0

    while count < total:
        try:
            review_batch, _ = reviews(
                app_id,
                lang="id",
                country="id",
                sort=Sort.NEWEST,
                count=15000,
                filter_score_with=None
            )
            for review in review_batch:
                if review["at"] >= cutoff_date:
                    all_reviews.append(review)

            # Stop jika review sudah melewati batas waktu
            if review_batch[-1]["at"] < cutoff_date:
                break

            count = len(all_reviews)
            print(f"{app_name}: {count} reviews from April 2022 and later...")

        except Exception as e:
            print("Error:", e)
            break

    df = pd.DataFrame(all_reviews)
    df['app_name'] = app_name
    return df

# Gabungkan semua review dari semua aplikasi
final_df = pd.DataFrame()
for name, app_id in apps.items():
    df = get_reviews(name, app_id, total=10000)
    final_df = pd.concat([final_df, df], ignore_index=True)

# Hapus kolom yang tidak perlu
columns_to_drop = [
    'reviewId', 'userImage', 'thumbsUpCount',
    'reviewCreatedVersion', 'replyContent',
    'repliedAt', 'appVersion'
]
final_df = final_df.drop(columns=columns_to_drop, errors='ignore')

# Simpan ke CSV
final_df.to_csv('playstore_reviews.csv', index=False)
print("Selesai! Komentar dari 1 April 2022 ke atas disimpan.")
```

```python


print("\nJumlah baris per aplikasi:")
for app_name in apps:
    app_reviews = final_df[final_df['app_name'] == app_name]
    num_reviews = len(app_reviews)
    print(f"{app_name}: {num_reviews} baris")
```
